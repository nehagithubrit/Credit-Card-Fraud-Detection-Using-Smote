# 1. **Tukey's Fences:**
Another method based on the IQR, but with different boundaries.

Calculate Q1 and Q3.

Q1, Q3
Determine the outlier boundaries.

Lower Fence = Q1 - k * IQR
Upper Fence = Q3 + k * IQR
Typically, k = 1.5 or 3.

Identify outliers: any data point outside the fences.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("/content/drive/MyDrive/PS_20174392719_1491204439457_log.csv")

# Define a function to detect outliers using Tukey's method for a given column
def detect_outliers_tukey(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers

# Apply the function to the 'amount' column
outliers_amount = detect_outliers_tukey(data['amount'])

# Calculate the number of outliers before removal
num_outliers_before = outliers_amount.sum()

# Remove outliers
cleaned_data = data[~outliers_amount]

# Calculate the number of outliers after removal
num_outliers_after = num_outliers_before - cleaned_data.shape[0]

# Print the number of outliers before and after removal
print("Number of outliers present before removal:", num_outliers_before)
print("Number of outliers removed:", num_outliers_after)

# Create a bar graph
labels = ['Outliers Before Removal', 'Outliers Removed']
values = [num_outliers_before, num_outliers_after]

plt.figure(figsize=(8, 6))
plt.bar(labels, values, color=['blue', 'orange'])
plt.title('Number of Outliers Before and After Removal')
plt.ylabel('Number of Outliers')
plt.show()
```
